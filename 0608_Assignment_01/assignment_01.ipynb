## PART 2: DESIGN YOUR TEAM

You will also be responsible for designing a pokemon team class. This class will be responsible for choosing which pokemon is sent into battle given the opponents pokemon. At the beginning of each round, a coin is flipped. Whoever wins the flip puts their pokemon on the field and attacks first. Whoever loses the flip, get's to choose their pokemon knowing who is already on the field. When a pokemon is knocked out, their trainer chooses which pokemon goes out next.

A trainer can also swap pokemon before a pokemon faints. ***A pokemon gains 5 hp per turn that it's on the bench (if it hasn't fainted)*** but swapping can only be done during the attack phase and makes you skip your attack.


***Hint: Make sure you aren't sending a fainted Pokemon out onto the field!***

Make sure you implement the following in your pokemon team class. You may add extra functions / attributes as you see fit but ensure that the following calls do not throw an error:

```python

team.team_name

team.roster
team.current_pokemon
```

````python
class PokemonTeam: # don't rename this class name

    def __int__(self):

        # IDENTITY ==========================================================
        self.team_name = "Team Design Media 577" # add your team name here
        self.trainer_name = "Thomas Gaudin" # Add your real name here
        self.salutation = "WATCH OUT IT'S TEAM DM577, AND WE WILL DESIGN YOUR DOWNFALL!"
        
        # ROSTER ============================================================
        # Make sure you are passing in your own custom classes that you made before
        # The roster is a dictionary with keys being the pokemon's names and the values
        # being an instantiation of 1 of each of the 3 pokemon classes your designed.
        self. roster = {
            "wigglyMon" : Pokemon1(), # replace with your first pokemon's name
            "coolDude" : Pokemon2(), # replace with your second pokemon's name
            "Frank" : Pokemon3() # replace with your third pokemon's name
        }

        # CURRENT POKEMON ==============================================================
        # The pokemon that is currently on the field. You will set this with the below functions.
        # Do not set this attribute value here

        self.current_pokemon = None # This value will be of type Pokemon1, Pokemon2, or Pokemon3


    # METHODS ======================================================================================
    
    def pick_on_go_first(self, opponent_team):
        """Pick the pokemon if you are going first / don't know the opponents first pokemon

        Args:
            opponent_team (PokemonTeam): The opponents pokemon team
        """

        # insert your logic here! For example you should probably analyze the pokemon types 
        # that your opponent has. 
        # You could also look at HP / Defense / whatever you want!! Be strategic.
        oppo_types = []

        for key in opponent_team.roster:
            oppo_pokemon = opponent_team.roster[key]
            oppo_types.append(oppo_pokemon.type1)
            oppo_types.append(oppo_pokemon.type2)

        if "snake" in oppo_types:
            self.current_pokemon = self.roster["wigglyMon"]
        
        elif "fire" in oppo_types:
            self.current_pokemon = self.roster["coolDude"]
        
        else:
            self.current_pokemon = self.roster["Frank"]
        
    def pick_on_go_second(self, opponent_team):
        """Pick the pokemon if you are going second. You get to see whoever is already on the field.

        Args:
            opponent_team (PokemonTeam): The opponents pokemon team
        """

        # you probably want to gauge this on who is currently on the field
        on_the_field = opponent_team.current_pokemon

        if on_the_field.type1 == "snake":
            self.current_pokemon = self.roster["Frank"]
        
        else: 
            self.current_pokemon = self.roster["wigglyMon"]
    
    def pick_on_faint(self, opponent_team):
        """Pick a new pokemon from your team if your current pokemon fainted.

        Args:
            opponent_team (PokemonTeam): The opponents pokemon team
        """

        # you probably want to gauge this on who is currently on the field
        # also make sure you aren't picking a fainted pokemon!
        on_the_field = opponent_team.current_pokemon

        if on_the_field.type1 == "snake" and self.roster["Frank"].hp != 0 :
            self.current_pokemon = self.roster["Frank"]
        
        elif self.roster["coolDude"].hp != 0:
            self.current_pokemon = self.roster["coolDude"]
        
        else:
            self.current_pokemon = self.roster["wigglyMon"]
        

    def switch_pokemon(self, opponent_team):
        """Switch out the current pokemon. This will run at the beginning of each your attack turns.
            You do not need to switch the pokemon if you don't want to.

        Args:
            opponent_team (PokemonTeam): The opponents pokemon team
        """

        # for example, the current pokemon is healthy - leave it in
        if self.current_pokemon.hp / self.current_pokemon._max_hp >= 0.6:  # pyright: ignore  <--- This is just put in here to ignore the type error. It will still work
            return
        
        # or switch to Frank... You probably want much more sophisticated logic
        self.current_pokemon = self.roster["Frank"]
````
