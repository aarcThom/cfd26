# Using Operators

The type of data we are dealing with determines what ***operators*** can act on it. An operator is a symbol denotes an action that can be performed on a data object.

### A quick note on **Immutability**

Here are a few of the operators we can use on a string. Note that when we use the following operators on strings, we are creating a new string. When working with bools, strings, floats, or integers, we always create new objects - we actually can't change existing strings or numbers. This is called ***immutability***, or we can say a string is ***immutable***.

## Using Operators with Strings

***Addition Operator*** ```+``` to create a new string whose values is the joined values of the input strings.

````python
# This expression concatenates two strings with the addtion operator
'hello ' + 'world'
````

***Multiplication Operator*** ```*``` to repeat strings

````python
# This expression repeats a string with the multiplication operator
'hello ' * 4 + 'world'
````

Note that standard order of operations applies in Python. ie. BEDMAS.

We can also use the ***slice operator*** ```[]```, to access a character at a certain location (index) in a string.

***Indices in Python ALWAYS START AT ZERO***

````python
# grab the first character in a string
'hello world'[0]
````

````python
# grab the third character in a string
'hello world'[2]
````

````python
# grab the last character in a string - we'll return to why this works when we look at lists
'hello world'[-1]
````

We can use the ***slice range operator*** ```[:]``` to access a range of characters in a string.

````python
# grabbing the first 4 characters - slice goes from start index to the index BEFORE the second number
'hello world'[0:4]
````

Not all operators are supported for a given type of object. For example, we can use the divsion operator ```/``` for number types, but string objects do not implement the ```/``` operator.

````python
'hello' / 'world'
````

# Variables

When programming, you will need to keep track of all the data that you are creating! We can assign a specific value (like a number or a string) to a variable in order to:

* Give the value a human-readable and descriptive name
* Refer to the value / variable in multiple places in our program
* Store the value for later use
* Reassign another value to a given variable when we need the data to update

We use the syntax ```variable_name = object```. The variable name always goes on the left-hand side of the equals symbol, and the value always goes on the right.


````python
#initialize a variable AND assign a string to it
dog_name = 'iggy'

#initialize a variable AND assign a string to it
greeting = 'hello '

#we can now interact with the variables just like we would with the objects themselves
greeting + dog_name
````

````python
# initialize a variable and assign an integer to it
big_number = 100000

# initialize a variable and assign an integer to it
biggish_number = 10000

# initialize a variable and assign a string to it
greeting = "hola "

# use the variables in the same way we use the data objects themselves

greeting * (big_number // biggish_number)
````

When we assign a data object such as a string or number to a variable, we are not permanently entangling the data the variable. Rather, I like thinking of *variables as pointing to a specific piece of data.* 

Allen Downey, in *Think Python* refers to a variables state - same thing. At a specific time in the code's execution a variable's state can be pointing to a certain value. We can represent this with a *state diagram*.

<img src="img\variables.png" width="600"> 

With this in mind, we can easily change the value that a variable is pointing towards!

````python
# initialize an a variable and assign a string to it
dog_name = "benny"

# replace the string that the variable is point to
dog_name = "iggy"

dog_name
````

Finally, variables in Python don't care about what *type* of object they are referring to. Because of this property, Python is considered a ***dynamically typed*** language.

This is in contrast to a ***statically typed*** language in which you need to declare the variable type before assigning an object or value to said variable. For instance in C# you would write this to create a variable that holds a string.

```string dog_name = "iggy";```


````python
#in Python, we don't need to specify the variable type. We can even change variable types on the fly (dynamically)
dog_name = 'iggy'

dog_name = 12345

dog_name
````

### Defining Functions

If we have a series of actions that we will be reusing, it's best practice to *encapsulate* that functionality in a function.

A **function definition** specifies the name of a new function and the sequence of statements that run when the function is called. We use the key `def` followed by a function name, parentheses `()`, and a colon `:` to create a function definition.

## Parameters

We can define ***parameters*** that can be used in our function. This allows use to apply the same calculations to different inputs.

We do so by listing the parameters' names within the parentheses seperated by comma(s) `,`.

````python
def average_number(num_1, num_2):
    avg_number = (num_1 + num_2) / 2
    print(avg_number) 
````

````python
average_number(10,6)
````

````python
average_number(12,44)
````

There is no restriction on the amount of parameters you can use in a function. For example, 0 parameters:

````python
def static_greeting():
    print("hello")

static_greeting()
````

1 parameter

*Note in the below example, we are using f-strings*

*F-strings* are a useful syntax for inserting variables or values within a string.

F-strings are expressed in the format `f"{<variable_name>} normal string"`

````python
def greet_name(name):
    print(f"hello {name}")
    
greet_name("Iggy")
````

Multiple parameters:

````python
def greet_names_several_times(name1, name2, greeting_amount):
    message = "hello " * greeting_amount + f"{name1} and {name2}!"
    print(message)

greet_names_several_times("Iggy", "Benny", 4)
````

We can also pass a variable amount of arguments if we don't know in advance how many arguments we want to pass into the function. To do so we add and asterisk `*` before the parameters name.

*Note that we haven't cover the 'for in' syntax yet. We will get to this later but right now I just want to demonstrate that we can pass an unknown amount of parameters*

````python
def greet_x_amount_of_names(*names):
    name_string = ""
    for name in names:
        name_string = name_string + f", {name}"
    
    print("Hello" + name_string)

greet_x_amount_of_names("iggy", "benny", "thomas", "hanna", "max")
````

Finally we can define optional argurments for our functions. To do so we use the regular syntax for creating a function, but we add an `= default_value` behind any optional argument. We must also make sure that our optional arguments come last in our arguments. For example,

````python
# a function with one optional argument
def pizza(topping1, topping2="cheese"):
    print(f"I've baked a {topping1} and {topping2} pizza!")

pizza("pepperoni")

pizza("veggies", "vegan cheese")
````

### Doc strings

## Scope

Scope is all about where you can use your variables and functions. It defines the areas in your code where certain variables are accessible. There are usually two main types: global scope, where a variable can be used anywhere in the program, and local scope, where a variable is only usable within a specific function or block. Scope can be nested.

Python deonotes changes in scope with indentation. Note how the indentations relate to different scopes.

````python
global_var_name_1 = "Iggy"
global_var_name_2 = "Benny"

def print_greeting(name1, name2):
    print(f"Hello {name1} and {name2}")


def calc_nums(num1, num2):
    calcd_num = num1 * num2 / 25
    return calcd_num

print_greeting(global_var_name_1, global_var_name_2)

returned_to_global_num = calc_nums(25, 100)
````

<img src="img\scope.png" width="600"> 

If a variable is created in a local scope, it is inaccesible outside of it's local scope.

````python
# when a variable is used within a function, it does not exist outside of a function - local scope
def fun_function():
    local_message = "hello"

print(local_message)
````

````python
# this rule also applies to parameters which are declared at the creation of a function

def greeting(message):
    print(message)

greeting('hello world')

print(message)
````

In Python, the opposite is not true. A variable defined in the main scope of code, can be used in functions or other indented blocks of code. This is called global scope.

````python
global_var = 'hello world!'

def greeting():
    print(global_var)

greeting()
````

It is generally not good practice to use global variables within functions. This can lead to confusing code quite quickly! Instead, it is best practice to pass variables into a function through parameters.

````python
global_var = 'hello world!'

def greeting(local_var): 
    print(local_var)

greeting(global_var) # the global variable is passed into the parameter / local variable 'local_var'
````

## Return Values

So far all the functions we have looked at have printed to the screen. We have't actually outputted objects from our functions. ***Return*** means to output an value from the functions local scope, to an outer (or global) scope. 

Let's see what happens if we try to assign the output of the function `average_number` to a variable.

````python
averaged_num = average_number(10, 6)
````

````python
print(averaged_num)
````

You can see that the value that the function evaluates does not get passed out of the function itself. It does get printed to the screen since we used `print()` but we can't actually use the calculated value in further calculations. We can see this behaviour more explicitly if we try to act upon a local variable defined within a function.

````python
def create_value():
    value = "hello"

create_value()
print(value)
````

We can use a `return` statement to assign the returned value to a variable defined in an outer scope.

````python
def repeat_greeting(num, greeting):
    repeated_greeting = greeting * num
    return repeated_greeting
````

````python
repeat_greeting(10, "hello ")
````

````python
# we can assign the output to a variable
big_greeting = repeat_greeting(5, "hola ")
````

````python
big_greeting
````

Note that when we use the **return** keyword the function instantly exits.

````python
def premature_exit(name_1, name_2):
    print(name_1)
    return
    print(name_2) # this name will never be printed
````

````python
premature_exit("frank", "lucy")
````
