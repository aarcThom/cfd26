## What is a Terminal?

A **terminal** (also called a terminal emulator) is a program that provides a text-based interface for interacting with your computer. Think of it as a window into the command-line world of your operating system.

In the early days of computing, terminals were physical devices—monitors and keyboards connected to a central computer. Today, we use software that mimics this behavior, letting us type commands instead of using a mouse and graphical interface.

<img src="img\terminal.png" width="600">  

On Windows, common terminal programs include:

* **Command Prompt (cmd)**
* **PowerShell**
* **Windows Terminal** (modern option that can run multiple terminal types)

On Mac and Linux, the built-in app is usually called **Terminal**.

---

## What is a Shell?

A **shell** is the program that actually interprets and executes the commands you type in a terminal. While a terminal is just the *window* where you interact, the shell is the *engine* that processes your commands.

Think of it this way:

* **Terminal** = The TV screen
* **Shell** = The cable box that processes and delivers the content

Common shells include:

* **Bash (Bourne Again Shell)** — The most common shell on Linux and Mac. Very powerful and widely used.
* **Zsh (Z Shell)** — Popular on Mac by default, offers additional features and customization.
* **PowerShell** — Microsoft's shell, designed for system administration and automation.
* **CMD (Command Prompt)** — The older Windows shell, simpler but less powerful than PowerShell.

When you open a terminal window, you're usually running a default shell. You can verify which shell you're using by typing:

```bash
echo $SHELL
```

or on Windows:

```powershell
echo $SHELL
```

---

## Why Do We Use Them?

While graphical user interfaces (GUIs) are intuitive, terminals and shells offer:

* **Speed** — Typing commands can be faster than clicking through menus, especially for repetitive tasks.
* **Automation** — Scripts can automate complex or repetitive workflows.
* **Power** — Many system tools and options are only available via the command line.
* **Precision** — You can specify exactly what you want to happen, without the ambiguity of a GUI.

In programming and AI contexts, you'll frequently need to use the terminal to install packages, run scripts, and manage your projects.

## Opening a Terminal (Mac)

The easiest way to open a terminal in a specific folder is to right-click directly in that folder:

1. Open **Finder** and navigate to the folder you want to work in.
2. **Right-click** (or Control-click) on the folder or inside it.
3. Select **"New Terminal at Folder"** from the context menu.

> **Note:** If you don't see this option, you may need to enable it. Go to **System Settings → Privacy & Security → Developer Tools** or enable it via **System Settings → Desktop & Dock → Turn on right-click menu option for New Terminal at Folder**.

The terminal will open with its working directory already set to that folder — so you're ready to run commands right away.

---

## Opening a Terminal (Windows)

The easiest way to open a terminal in a specific folder is to right-click inside that folder in File Explorer:

1. Open **File Explorer** and navigate to the folder you want to work in.
2. **Right-click** on an empty area inside the folder (not on a file).
3. Select **"Open in Terminal"** from the context menu.

> **Note:** This opens **Windows Terminal** (or PowerShell on older systems) with the working directory set to that folder. If you don't see "Open in Terminal", make sure **Windows Terminal** is installed — you can get it free from the Microsoft Store.

The terminal will open already pointing to that folder — no need to navigate with `cd`.

### Reading the Prompt

Before you type a command, the terminal shows a **prompt** — a line of text that tells you where you are and signals that the shell is ready for input. It typically looks something like this:

```
yourname@MacBook projects
```
or on Windows:
```
yourname@DESKTOP /c/Users/yourname/projects
```

---

### Where am I? — `pwd`

`pwd` stands for **print working directory**. It shows your current location in the file system.

```bash
pwd
```

Example output on Mac: `/Users/yourname/projects`  
Example output on Windows: `/c/Users/yourname/projects`

---

### What's in this folder? — `ls`

`ls` lists the files and folders in your current directory.

```bash
ls
```

On Mac, add `-l` for a detailed list, or `-a` to show hidden files (files starting with `.`):

```bash
ls -l
ls -a
```

On Windows to show hidden files
```bash
ls -Hidden
```

---

### How do paths work?

A **path** describes a location in the file system. There are two kinds:

* **Absolute path** — starts from the root of the file system. Always begins with `/`.
  ```
  /Users/yourname/projects/myapp
  ```

* **Relative path** — starts from wherever you currently are.
  ```
  projects/myapp
  ```

Two special shortcuts:
* `.` — the current directory
* `..` — the parent directory (one level up)

---

### Moving around — `cd`

`cd` stands for **change directory**.

```bash
cd projects          # go into a folder called 'projects'
cd projects/myapp    # go into a nested folder
cd ..                # go up one level
cd ../..             # go up two levels
cd ~                 # go to your home directory
```

---

### Creating a folder — `mkdir`

`mkdir` creates a new directory.

```bash
mkdir myfolder
```

Use `-p` to create nested folders in one command:

```bash
mkdir -p projects/myapp/src
```

---

### Deleting files and folders — `rm`

`rm` removes files. **This is permanent — there is no Recycle Bin.**

```bash
rm myfile.txt          # delete a file
rm -r myfolder         # delete a folder and everything inside it
```

> **Warning:** `rm -r` is irreversible. Double-check your path before running it.

## LLMs, Agents, and Harnesses - Using OpenCode

### LLMs

You have probably interacted with an **LLM** (large language model) before. An LLM such as chatGPT, Gemini, or Claude is a generative next-token prediction model. In other words, when you type in a sentence and LLM will predict what components of words statistically will come next.

For instance,

***THE DOG RAN DOWN THE***

may be a sentence which most likely ends in ***street.*** This is essentially how LLMs can generate code. Given a question, like "how do I print hello world in Python", the LLM will continue guessing the next word until it reaches a special termination character. This is very simplified. If you're more interested in a slightly deeper dive [check out this link](https://medium.com/data-science-at-microsoft/how-large-language-models-work-91c362f5b78f).

LLMs are stateless. In otherwords, they don't actually remember things. Each time you send a request to an LLM, you are actually sending your entire conversation history. This is what creates the illusion of continuity. The LLM just keeps having more context to guess what words come next!

### Agents

An agent is simply a loop of LLM calls. This loop is often called the **ReAct loop** (short for **Reasoning + Acting**). An agent can call tools and insert the output of these tools into it's building conversation context.

<img src="img\react_agent.png" width="600">

#### What Are Tools?

Tools are functions the agent is allowed to call. Common ones for a coding agent include:

* **read_file** — read the contents of a file
* **write_file** — create or overwrite a file
* **run_command** — execute a shell command and get back the output
* **search** — search the codebase for a pattern or keyword
* **web_fetch** — retrieve content from a URL

The agent doesn't run code itself — it asks the tool to do it, then reads what came back.

---

### Harnesses

A harness is the program in which the looped LLM calls are run. This is the program that allows the LLM to call tools such as the weather API call made in the illustration above.

<img src="img\coding_harness.png" width="600">  

Using a harness such as *Claude Code*, *Open Claw*, *OpenCode*, *Codex*, etc. allows us to have an LLM use write tools within our codebase to edit or generate our code. We can also ask an LLM to explain the contents of a codebase with read tools. The agent actually knows how to access the files in question with search tools, and can call on extra information using web_fetch tools!

---

#### A Concrete Example

> "Add a function that reverses a string to `utils.py`"

1. Agent reads `utils.py` to understand what's already there
2. Agent writes an updated version of `utils.py` with the new function
3. Agent runs the test suite to check nothing broke
4. Agent reads the test output
5. If tests pass → done. If not → agent reasons about the failure and tries again.

---

#### Why This Matters

Because the agent is looping and observing results, it can recover from mistakes, adapt to what it finds, and handle tasks that can't be solved in a single step. It's not just predicting text — it's using feedback from the real environment to make progress.

## Installing OpenCode (Mac)

curl comes pre-installed on Mac, so you can install OpenCode in one command. Open a terminal and run:

```bash
curl -fsSL https://opencode.ai/install | bash
```

This downloads the official install script and runs it. It will detect your system, download the correct binary, and add `opencode` to your PATH automatically.

Once it finishes, close and reopen your terminal, then verify the install:

```bash
opencode --version
```

---

## Installing OpenCode (Windows)

Windows doesn't have a built-in package manager like curl for installing CLI tools, so we'll use **Scoop** — a lightweight package manager for Windows that installs programs without needing admin rights or UAC popups.

### Step 1: Install Scoop

Open **PowerShell** (not Command Prompt) and run these two commands in order:

```powershell
Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser
```

This allows PowerShell to run installation scripts. Then:

```powershell
Invoke-RestMethod -Uri https://get.scoop.sh | Invoke-Expression
```

Scoop will install itself to your home directory. Once it's done, close and reopen your terminal.

### Step 2: Install OpenCode

With Scoop installed, run:

```powershell
scoop install opencode
```

Scoop handles downloading the binary and adding it to your PATH automatically.

### Step 3: Verify and open OpenCode

run:

```powershell
opencode
```
