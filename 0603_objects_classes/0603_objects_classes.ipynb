## Objects Everywhere!

Every piece of data in Python is an ***object***. We have been working with objects this entire time! Strings are objects, numbers are objects, lists and dictionaries are objects!

Each object has:

* An ***Identity*** (A unique identifier for the object in memory — accessible via `id()`)
* A ***Type*** (What kind of object is it? — accessible via `type()`)

Objects will also usually have:

* ***Methods*** (Sequences of commands the object can perform on itself or other objects)
* A ***Value*** (For simple built-in types like `int`, `float`, `bool`, and `str`: the actual data represented — the number, the text, etc.)

Custom Objects we define will also usually have:

* ***Attributes*** (Descriptors of the object and extra stored values)

We can act upon objects with:

* ***Functions*** (Sequences of commands applied to an object from outside — e.g. `len(x)`, `print(x)`)
* ***Operators*** (Symbols that manipulate objects or produce new ones — e.g. `+`, `==`, `[]`)

## Custom Objects - Classes

***Classes*** let us defined our own custom object types and bundle related data and behavior into a single reusable unit.

The ***class*** is the blueprint for a type of object. We use this class to create ***instances*** of that class type. 

For instance, we can define a `Pokemon` type with a class. We can then create instances of the Pokemon type, such as `Charizard` or `Pikachu`.

## Creating a Pokemon Class

At the very least we need two things to create a class:

* A class definition `class Classname:`
    * This tells Python that we are defining a new class
    * It is conventional that class's start with uppercase letters
    * no parentheses are included in the class definition
* An initializer `def __init__(self):`
    * An initializer is the method called to create (initialize) and instance of the class
    * Usually *attributes* are set in the initializer
    * The keyword `self` means the instance.
    * We call `__init__` a *dunder method* (double underscore) method.

```python
class Pokemon:
    def __init__(self):
        pass # pass simply means that nothing happens while satisfying Python syntax rules - no empty methods
```

```python
pikachu = Pokemon()
```

```python
type(pikachu)
```

You can see that we have created a new class of the type Pokemon and have initialized an instance of that class. However our class doesn't really do much yet. Let's start by adding some attributes.

```python
class Pokemon:
    def __init__(self, name:str, dex_num:int, t1:str, t2:str) -> None:
        """Create a pokemon instance!

        Args:
            name (str): Pokemon type (ie. Bulbasaur)
            dex_num (int): Pokedex Number
            t1 (str): Type 1
            t2 (str): Type 2
        """
        self.name = name # note that attributes can have the same name as __init__ parameters
        self.pokedex_num = dex_num # or they can have different names!
        self.type1 = t1
        self.type2 = t2
```

```python
bulbasaur = Pokemon("bulbasaur", 1, "plant", "poison")
```

Now that we have some attributes, we can add some methods. **Methods** are functions contained within a class that an instance of that class can call. When we call methods from custom classes in Python, we must always include the first parameter `self`. 

We don't actually have to pass anything into `self` however. Python automatically passes the instance of the object. It is this behaviour that let's us access attributes within a method.

```python
class Pokemon:
    def __init__(self, name:str, dex_num:int, t1:str, t2:str) -> None:
        """Create a pokemon instance!

        Args:
            name (str): Pokemon type (ie. Bulbasaur)
            dex_num (int): Pokedex Number
            t1 (str): Type 1
            t2 (str): Type 2
        """
        self.name = name
        self.pokedex_num = dex_num
        self.type1 = t1
        self.type2 = t2

    def greet(self)->str:
        """Returns a greeting from the pokemon
        Returns:
            str: The pokemon saying hello
        """
        poke_greet = f"Hello, I am {self.name}. I wish you a {self.type1} filled day!"
        return poke_greet
```

```python
bulby = Pokemon("bulbasaur", 1, "plant", "poison")
```

```python
bulby.greet()
```

We can also use a method to update the Pokemon's attributes.  We can pass a value into a method to alter attributes in a controlled manner. 

Often times it's useful to signal that an attribute *shouldn't* be altered from outside an instance of a class. For instance, we want to record a Pokemon's max HP, and also there current HP. There is no real way to enforce an attribute being changed be something outside an object in Python, but by convention we mark attributes that shouldn't be messed with an underscore.

```python
class Pokemon:
    def __init__(self, name:str, dex_num:int, t1:str, t2:str, hp:int) -> None:
        """Create a pokemon instance!

        Args:
            name (str): Pokemon type (ie. Bulbasaur)
            dex_num (int): Pokedex Number
            t1 (str): Type 1
            t2 (str): Type 2
            hp (int): Hit Points
        """
        self.name = name
        self.pokedex_num = dex_num
        self.type1 = t1
        self.type2 = t2
        self.hp = hp # the public facing, alterable HP

        self._hp = hp # we use an underscore to denote that this attribute should NOT be altered outside the object

    def greet(self)->str:
        """Returns a greeting from the pokemon
        Returns:
            str: The pokemon saying hello
        """
        poke_greet = f"Hello, I am {self.name}. I wish you a {self.type1} filled day!"
        return poke_greet
    
    def heal(self, amount:int) ->None:
        """Boosts the pokemon's HP up to max HP. Print's a healing update.

        Args:
            amount (int): Amount to boost HP.
        """

        if amount <= 0:
            print(f"{self.name} felt no effect!")
            return

        if self.hp == self._hp:
            print(f"{self.name} is already at max HP. Nothing done.")
            return
        
        if self.hp + amount > self._hp:
            print(f"{self.name} gained {self._hp - self.hp} points!. They're at max health!")
            self.hp = self._hp
            return
        
        self.hp += amount
        print(f"{self.name} was healed {amount} points! They are at {self.hp} HP.")
        return
```

```python
bulby.heal(10)
```

Finally, we can implement our own *dunder methods* to define things like how operators act on an object type, or how those objects are represented.

For instance, our pokemon is currently not very descriptive.

```python
bulby
```

We can change this behaviour by implemented the `__repr__` dunder method, which changes how an object is represented in the notebook's output.

```python
class Pokemon:
    def __init__(self, name:str, dex_num:int, t1:str, t2:str, hp:int) -> None:
        """Create a pokemon instance!

        Args:
            name (str): Pokemon type (ie. Bulbasaur)
            dex_num (int): Pokedex Number
            t1 (str): Type 1
            t2 (str): Type 2
            hp (int): Hit Points
        """
        self.name = name
        self.pokedex_num = dex_num
        self.type1 = t1
        self.type2 = t2
        self.hp = hp # the public facing, alterable HP

        self._hp = hp # we use an underscore to denote that this attribute should NOT be altered outside the object


    def __repr__(self) -> str:
        return f"Pokemon Type: {self.name}\nHP: {self.hp}/{self._hp}"

    def greet(self)->str:
        """Returns a greeting from the pokemon
        Returns:
            str: The pokemon saying hello
        """
        poke_greet = f"Hello, I am {self.name}. I wish you a {self.type1} filled day!"
        return poke_greet
    
    def heal(self, amount:int) ->None:
        """Boosts the pokemon's HP up to max HP. Print's a healing update.

        Args:
            amount (int): Amount to boost HP.
        """

        if amount <= 0:
            print(f"{self.name} felt no effect!")
            return

        if self.hp == self._hp:
            print(f"{self.name} is already at max HP. Nothing done.")
            return
        
        if self.hp + amount > self._hp:
            print(f"{self.name} gained {self._hp - self.hp} points!. They're at max health!")
            self.hp = self._hp
            return
        
        self.hp += amount
        print(f"{self.name} was healed {amount} points! They are at {self.hp} HP.")
        return
```

## Making tons of Pokemon!
Remember that Pokemon JSON file we had? We can now make ALOT of pokemon instances.

```python
import json

with open("data/pokemon.json", mode="r", encoding="utf-8") as context_manager:
    pokemon_dicts = json.load(context_manager)

pokemon_dicts
```

```python
pokemon = []

for name_key in pokemon_dicts:
    # grab the needed values from the dictionary
    name = name_key # remember that the keys were the names
    pokedex_num = pokemon_dicts[name]["pokedex_number"]
    type_1 = pokemon_dicts[name]["type1"]
    type_2 = pokemon_dicts[name]["type2"]
    hp = pokemon_dicts[name]["hp"]

    # initialize the new pokemon object and add to the list
    pokemon.append(Pokemon(name, pokedex_num, type_1, type_2, hp))
```

```python
# pikachu is pokedex # 25
pikachu = pokemon[24] # indices start at 0

pikachu
```

```python
pikachu.greet()
```
