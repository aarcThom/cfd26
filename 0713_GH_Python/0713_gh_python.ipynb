### Breakpoints and the Breakpoints Tab

Using Python notebooks is a great way to learn coding as it allows us to run our code in small cells. In the Grasshopper code editor, more akin to a traditional code editor, we will write our code in one longer document. To help us troubleshoot our code in smaller sections, we will use ***breakpoints***.

***Breakpoints*** denote lines that the editor will pause on during runtime. They allow us to examine variable values at given points in time or certain iterations in a loop.

#### Adding and Removing Break Points
To add a break point to your code, click to the left of the line number. You will see a red dot appear next to the line. To remove a break point, you click the red dot.

As shown in the animation above, if you click the green run arrow (now with a little bug), you will start your code. The arrow then changes to a double arrow, that allows your to push the code forward one iteration to the breakpoint. In other words, you will not pause on the lines of code between the break points. If you open the *variables tab*, you will be able to see the value of variables at each point in running the code through the breakpoint.

#### Iterating Through the Code Line by Line

### List Access, Item Access

In Python, we are used to interacting with collections of objects through lists. Grasshopper offers us 3 different ways to interact with input data.

#### Item Access
If you were to write a list of names on a GH panel, and input them into a GH input / python variable `names` and iterate through them as we have learnt to in this course, you might get a confusing result.

<img src="img/09_item_access_A.png" width="1000"> 

The reason this behaviour is happening, is that Grasshopper Python components default to passing data from the GH input to the Python variable *per item*. This is called *item access*. You can switch a component to and from item access by right clicking the input.

<img src="img/09_item_access_B.png" width="800">

With *item access*, Grasshopper will not consider collections of objects. It will literally run the written code on each item passed through the input. In this example, the list of names is not considered as a list - rather *each name is iterated through as a string*.

In other words, the first time this component runs:

```
names = "timmy"

for name in names:
    print(f"Hello {name}!")

output:
Hello t!
Hello i!
Hello m!
Hello m!
Hello y!

```

The component will then run the same iteration on each name in isolation.

### What are Libraries?

Now, we want to work with Rhino geometry in the same way we worked with these primitive types. To do so, we will be working with a ***library*** called **RhinoCommon**.

A library is a pre-coded collection of functions and objects. We have already worked with the library, `random`.

A library is organized into  packages and modules. Often, the term library is used generically to mean a collection of reusable code. A library can be used to either describe a collection of packages which each contain their own collection of modules, OR a library is frequently used interchangeably with package, as in, a library is a collection of modules.

<img src="img//11_libraries.png" width="800">

Regardless of the way the term library is defined in a particular context, just keep in mind that we are referring to a system of organizing and accessing pre-written code.

For an in-depth explanation of library / package / module organization, see this [link](https://learnpython.com/blog/python-modules-packages-libraries-frameworks/).

Another term you will encounter while referencing the documentation for Rhino libraries is *namespace*. In different programming languages, similar systems and functionalities are given different names. Namespace is how you would refer to libraries, packages, and modules if you were programming in C# or C++. The documentation was originally written for those languages and, as such, uses this terminology. In this class, we will stick with the Python terminology.


### Importing Libraries

If you wanted to import the library `Random`, you would do so at the very beginning of your code with this statement: `import random`

Here are a few notes about library importing syntax:

* The library import statement always starts with import as in `import random`
* As we mentioned, Libraries may further contain packages and modules. You access these packages and modules using dot notation.
    * For example, the Rhino library contains packages such as geometry, document, collections.
    * To access the geometry package, we would type: `import Rhino.Geometry`
    * To access the collections module inside the geometry package, we would type: `import Rhino.Geometry.Collections`
* You may specify an *alias* for a library upon import. 
    * The syntax for this statement would be `import library_name as ln` or whatever abbreviation you’d like.
    * This allows you to reference the library with a shorthand alias through your code. You will see that library names with their modules can get quite long so it’s common practice to shorten their names.
    * For example, we could write: `import Rhino.Geometry as rg`
* You will often import multiple libraries. These import statements will come one line after another at the top of your code.

### Introducing the RhinoCommon Library

In this course we will focus on using the library [RhinoCommon](https://developer.rhino3d.com/api/rhinocommon/) for Grasshopper / Rhino scripting.

* RhinoCommon is Rhino’s main cross-platform SDK (software development kit). See the [official introduction here](https://developer.rhino3d.com/guides/rhinocommon/what-is-rhinocommon/).
* It is extensively documented and offers the programmer access to all functionality available in Rhino3D.
* RhinoCommon is imported in Python using the word `Rhino`, as in: `import Rhino`
* RhinoCommon contains the package `Geometry` that contains all geometry available in Rhino and their corresponding methods. You would import this package with: `import Rhino.Geometry`
* RhinoCommon also contains packages for document handling, UI, and geometry representation among other things.

RhinoCommon is the most comprehensive, and most actively maintained library for dealing with all things Rhino. There are however two other libraries that you might see referenced online. I list them here to avoid confusion in self research - do not use these libraries for any of this course's work. 

#### RhinoScriptSyntax
* A wrapper over the Rhino.Geometry library which provides similar commands and functionality to Rhinoscript
* A wrapper refers to code that has been formatted in a way that another language or program can use it. For example, a Javascript library might have a Python wrapper that allows its functionality to be accessed via Python.
* In this case, rhinoscriptsyntax is a wrapper that was designed to allow Rhino’s proprietary scripting language RhinoScript to use RhinoCommon.
* This wrapper was brought into Python to allow coders experienced with RhinoScript to transition to Python.

#### ghpythonlib
* This library contains the package ghpythonlib.components which  is also referred to as ‘node in code’ as it allows you to implement the functionality of familiar Grasshopper components into your Python code.
* See the [official introduction page](https://developer.rhino3d.com/guides/rhinopython/ghpython-call-components/) for a good introduction.
* This library is useful for quickly sketching out ideas or super-charging GH components’ functionalities with recursive behaviours.
* This library also contains a useful package, ghpythonlib.treehelpers which allows you to quickly convert Python lists to Grasshopper trees and vice versa. 
* Ghpythonlib is limited in scope and capabilities in comparison to RhinoCommon.

***Again, I mention these libraries only to avoid confusion that might arise during self-research. To make life easier for yourself, stick solely to RhinoCommon.***


## RhinoCommon.Geometry - Object Methods and Attributes

Bookmark the page **https://developer.rhino3d.com/api/rhinocommon/** NOW. 

You will have it open constantly while programming in Grasshopper. Notice on the left hand side the `Geometry` package - click it to display the package info in the right panel, then click the arrow to expand the package's contents.

<img src="img/12_rhino_common_A.png" width="800">

Now, scroll down, in the left panel, until you reach the module `Sphere`. Click it to expand. You will see that this module represents *an object Sphere*.

<img src="img/12_rhino_common_B.png" width="800">

See that it contains a few things:
* **A Constructor** or how we initialize a new sphere object
    * In contrast to a string which is initialize with double quotes as in `"string"`, we can see that we can initialize a new sphere object using the syntax `Sphere(equatorialPlane:Plane, radius:float)` or `Sphere(center:Point3d, radius:float)`
    * All Rhino geometrical objects we work with will use this function-like syntax `Object(arg1, arg2, ...)`
    * Note that in the documentation, the syntax is slightly different - this is due to the fact that RhinoCommon was developed initially for the programming languages, C++ and C#. If you aren't sure what the exact syntax would be for an object, ask ChatGPT to translate from C# to Python for you!
* **Properties** aka *Attributes*
    * In our lecture on objects (`0515_objects_variables`) we talked about attributes - extra values an object can store. We mentioned how strings and numbers don't really haveuseful attributes, and that we'd revisit attributes at a later time. That time is now!
    * **Attributes** (called properties in the docs) are extremely important for custom objects. For instance, the attribute `Center` which contains an object of type `Point3d`, determines the location of a sphere. The attribute `Diameter` which contains a `float` type object, determines the sphere's diameter.
    * We access and set attributes using dot notation, but we do not include parentheses as we do in methods. For instance `print(object.attribute)` or `object.attribute = updated_value`
* **Methods**
    * Each geometry object in RhinoCommon will contain methods used to affect the position, or shape / size / orientation of a given object. In the case of the sphere object, these are called `Transform`, `Rotate`, and `Translate`.
    * Objects will also contain utitlity or information providing methods such as `ClosestPoint`, `NormalAt`, etc.


So, to tie back to an earlier lecture, you can see that each Rhino Geometry type fits the rules of objects we defined in the second class! Above we have an object of:
* Type = Sphere
* Value = The Sphere Rhino Geometry
* Attributes = Diameter, Center, etc.
* Methods = Transform, Rotate, ClosestPoint, etc.

We will also see that we can use functions on these objects, collect them in lists, and sometimes even use arithmetic operators on them (as in Point3ds).
