This Pokemon battle will be a simplified, and automated round between two teams of Pokemon - your team, and the computer's team. The flow of the programming will go as such:

* The computer chooses a team of three Pokemon, in the order that they battle.
* You choose a team of three Pokemon, in the order that they will battle.
* The Pokemon take turns attacking each other.
    * When a Pokemon is reduced to 0 or less `hp`, it is knocked out and replaced.
    * When all three pokemon are knocked out, the opposing team wins.
